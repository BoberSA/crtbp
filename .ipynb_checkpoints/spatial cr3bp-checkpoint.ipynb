```python
def omega(y):
    r1 = np.sqrt((y[:,0] + mu2)**2 + y[:,1]**2 + y[:,2]**2)
    r2 = np.sqrt((y[:,0] - 1 + mu2)**2 + y[:,1]**2 + y[:,2]**2)
    return 0.5*(y[:,0]**2 + y[:,1]**2) + (1 - mu2) / r1 + mu2 / r2
    
def jacobi_const(y):
    return 2*omega(y) - y[:,3]**2 - y[:,4]**2 - y[:,5]**2
```

```python
def dimless2km(y):
    ykm = y.copy()
    ykm[:,0] = (ykm[:,0] - L[1])*ME
    ykm[:,1] *= ME
    ykm[:,2] *= ME
    return ykm
```
