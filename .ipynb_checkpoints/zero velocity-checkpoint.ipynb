```python
def r1(x, y, z):
    global mu
    return np.sqrt((x+mu)**2 + y**2 + z**2)

def r2(x, y, z):
    global mu
    return np.sqrt((x-1+mu)**2 + y**2 + z**2)

def Omega(x, y, z):
    global mu
    return 1/2*(x**2+ y**2) + (1-mu)/r1(x, y, z)+mu/r2(x, y, z)
    
```

```python
def allowed_zone(x, y, z, C):
    return 2*Omega(x, y, z) - C
```
